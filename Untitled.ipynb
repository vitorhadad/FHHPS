```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use("ggplot")
```

# <center>FHHPS - Simulation results</center>

## Data:

Random coefficients: 

\begin{align*}
\begin{bmatrix}
A_{1}\\ B_{1}
\end{bmatrix}
\sim
\mathcal{N}\left(
\begin{bmatrix}
1 \\ 2
\end{bmatrix}, \
\begin{bmatrix}
2 & 1 \\ 1  & 2
\end{bmatrix} \right)
\end{align*}

Second period shocks:

\begin{align*}
\begin{bmatrix}
U_{2}\\ V_{2}
\end{bmatrix}
\sim
\mathcal{N}\left(
\begin{bmatrix}
.5 \\ .5
\end{bmatrix}, \
\begin{bmatrix}
1 & 0 \\ 0  & 1
\end{bmatrix} \right)
\end{align*}

Regressors (note the de-meaning step)

\begin{align*}
X_{1}^* &= .2A_{1} + .5B_{1} + .5A^2 - .3B^2 + \epsilon \qquad \epsilon \sim \mathcal{N}(0,5) \\
X_{1} &= X_{1}^* - E[X_{1}^*]
\end{align*}

\begin{align*}
X_{2} &=  \mathcal{N}(0,5) \\
\end{align*}

Regressand:

\begin{align*}
Y_{1} &=  A_{1} + B_{1}X_{1}\\
Y_{2} &=  (A_{1}+U_{2}) + (B_{1} + V_{2})X_{2}\\
\end{align*}

## Simulation results

<b>Step 1 parameters:</b>
<ul>
    <li>$b1 = 0.1$: Kernel parameter used in the estimation of shocks  
</ul>

<b>Step 2 parameters:</b>
<ul>
    <li>$b2 = 0.1$: Kernel parameter used in the estimation of random coefficients
    <li>$t = .2$: Lower bound on modulus of the denominator
    <li>$d = 1$: Keep only observations with $|X_{2} - X_{1}| > d$
    <li>$lim = .5$: Estimate characteristic function over square $[-lim, +lim]^2$
    <li>$n\_pts = 100$: Number of pairs over which characteristic function will be evaluated
</ul>

(1, x1)
(1, x2)

-->

1/(x2 - x1) * ()


```python
n_obs = [500, 2000, 5000, 10000, 20000, 50000]
```

```python
params = ["n_obs", "b1", "b2", "t", "d", "lim", "n_pts"]
coeffs = ["EU", "EV", "VarU", "CovUV", "VarV", "EA", "EB", "VarA", "CovAB", "VarB"]
colnames = params + coeffs
truth = pd.Series(dict(zip(coeffs,[.5, .5, 1, 0, 1, 1, 2, 2, 1, 2])))
colors = plt.cm.Dark2(np.linspace(0,1,len(n_obs)))
```

```python
ticks = {"EU":np.linspace(-.5, 1.5, 5),
        "EV":np.linspace(-.5, 1.5, 5),
        "VarU":np.linspace(0, 4, 5),
        "CovUV":np.linspace(-1, 1, 5),
        "VarV":np.linspace(0, 2, 5),
        "EA":np.linspace(0, 2, 5),
        "EB":np.linspace(1, 3, 5),
        "VarA":np.linspace(0, 4, 5),
        "CovAB":np.linspace(-1, 3, 5),
        "VarB":np.linspace(0, 4, 5)}
```

```python
for m, n in enumerate(n_obs):
    
    table = pd.read_csv("results_" + str(n) + ".csv", skiprows = 1, header = None).drop_duplicates().dropna()
    table.columns = params + coeffs

    fig, ax = plt.subplots(2,5, figsize = (30, 10))
    fig.subplots_adjust(wspace = .1, hspace = .5)
    fig.suptitle("N = " + str(n), fontsize = 30)

    for k, coeff in enumerate(coeffs):

        i,j = np.unravel_index(k, ax.shape)

        p = ax[i,j].hist(table[coeff], alpha = .6, 
                         bins = 20, normed = True,
                        color = colors[m,])
        p = ax[i,j].vlines(truth[coeff], 
                           ymin = ax[i,j].get_ylim()[0], ymax = ax[i,j].get_ylim()[1],
                           linewidth = 5, color = "black")
        p = ax[i,j].vlines(table[coeff].mean(), 
                           ymin = ax[i,j].get_ylim()[0], ymax = ax[i,j].get_ylim()[1], 
                           linewidth = 5, color = "red")
        p = ax[i,j].set_title(coeff, fontsize = 25)
        p = ax[i,j].set_xticks(ticks[coeff])
        p = ax[i,j].set_xticklabels(ticks[coeff], fontsize = 15)
        p = ax[i,j].set_xlim((ticks[coeff][0],ticks[coeff][-1]))
```

```python
for k, coeff in enumerate(coeffs):

    fig, ax = plt.subplots(1, 6, figsize = (30, 5), sharex = True, sharey = True)
    fig.subplots_adjust(wspace = .1, hspace = .5)
    fig.suptitle(coeff, fontsize = 30)

    for m, n in enumerate(n_obs):

        sims = pd.read_csv("results_" + str(n) + ".csv", 
                            skiprows = 1, header = None).drop_duplicates().dropna()

        sims.columns = params + coeffs
        sims = sims[coeff]

        p = ax[m].hist(sims, alpha = .8, normed = True, color = colors[m,])
        
        p = ax[m].set_title(str(n), fontsize = 25)
        p = ax[m].set_xticks(ticks[coeff])
        p = ax[m].set_xticklabels(ticks[coeff], fontsize = 15)
        p = ax[m].set_xlim((ticks[coeff][0],ticks[coeff][-1]))

    
    for m, n in enumerate(n_obs):

        sims = pd.read_csv("results_" + str(n) + ".csv", 
                            skiprows = 1, header = None).drop_duplicates().dropna()

        sims.columns = params + coeffs
        sims = sims[coeff]
        p = ax[m].vlines(truth[coeff], ymin = 0, ymax = ax[m].get_ylim()[1], linewidth = 2, color = "black")
        p = ax[m].vlines(sims.mean(), ymin = 0, ymax = ax[m].get_ylim()[1], linewidth = 2, color = "red")
        
```

```python
table = pd.DataFrame()
for n in n_obs:
    
    tmp = pd.read_csv("results_" + str(n) + ".csv", 
                      skiprows = 1, header = None,
                      names = colnames).drop_duplicates().dropna()
    tmpmean = tmp[coeffs].mean(axis = 0)
    tmpmean["n_sims"] = tmp.shape[0]
    
    diff = (tmp - truth)[coeffs]
    tmpmean["meanabsloss"]= np.mean(np.sum(np.abs(diff.values), axis = 1))
    tmpmean["meanrmsloss"] = np.mean(np.sqrt(np.sum(diff.values**2, axis = 1)))
    
    table = pd.concat([table, tmpmean], axis = 1)

table = table.T
table.index = [500, 2000, 5000, 10000, 20000, 50000]
table = table[coeffs + ["meanrmsloss", "meanabsloss"] + ["n_sims"]]

table
```
